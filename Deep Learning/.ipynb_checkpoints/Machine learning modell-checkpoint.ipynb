## Machine learning modell

```python
# Imports
import numpy as np
import pandas as pd
import sklearn.metrics as metrics
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
from tensorflow.keras.optimizers import Adam
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read Data
df = pd.read_csv('Pokemon-Daten.csv', sep=";")
display(df.head(10))
df.status.value_counts()
```

```python
# Rename Status 
df["status"] = df["status"].str.replace("Normal", "Non_Legendary")
df["status"] = df["status"].str.replace("Mythical", "Non_Legendary")
df["status"] = df["status"].str.replace("Sub Legendary", "Non_Legendary")

df.status.value_counts()
```

```python
df["sum"] = (df["hp"] + df["attack"] + df["defense"] + df["sp_attack"] + df["sp_defense"] + df["speed"])/6
#display(df["sum"])
```

```python
# Define X and Y 
x = df[["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]]
#x.drop(columns=["name", "german_name", "species", "type_1", "type_2", "generation", "type_number", "pokedex_number", "height_m", "weight_kg"], inplace=True)
#x = x.dropna()
y = df["status"]
#x.drop(columns=["status"], inplace=True)

print(x)
print(y)
```

```python
# Transform X and Y to Arrays
display(x.info())
```

```python
x = x.to_numpy()
y = y.to_numpy().reshape(-1,1)

display(x)
display(y)
```

```python
#np.nan_to_num(x, nan=0, posinf=100, neginf=0)
#array_sum = np.sum(x)
#array_has_nan = np.isnan(array_sum)

#print(array_has_nan)
```

```python
# Class Labelling
# Instead of having classes such as [1,2,3]
# It creates a output of size 3 flags, each flag represents one class
encoder = OneHotEncoder(sparse=False)
y_ = encoder.fit_transform(y)

# Data Split
X_train, X_test, Y_train, Y_test = train_test_split(x, y_, test_size=0.20)

display(X_train)
display(Y_train)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```

```python
# Create Model with 3 layers
model = Sequential()
model.add(Input(shape=(6), name='input'))
model.add(Dense(18, activation='relu', name='fc1'))
model.add(Dense(18, activation='relu', name='fc2'))
model.add(Dense(2, activation='softmax', name='output'))
```

```python
# Adam Optimizer
optmizer = Adam(learning_rate=0.001)
model.compile(optmizer, metrics=['accuracy'] , loss='categorical_crossentropy')
print('Summary of the model')
print(model.summary())
```

```python
# Fit Model
model.fit(X_train, Y_train, verbose=1, batch_size=5, epochs=200)
```

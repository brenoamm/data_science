```python
# Imports
import numpy as np
import pandas as pd
import sklearn.metrics as metrics
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
from tensorflow.keras.optimizers import Adam
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris Dataset
# Features = [Sepal Length,รถ Sepal Width, Petal Length, Petal Width]
iris_data = datasets.load_iris()

x = iris_data.data

#display(iris_data.target)
y = iris_data.target.reshape(-1,1)
#display(y)

target_names = list(iris_data.target_names)
print(target_names)
```

```python
#Plot Data
column_names = iris_data.feature_names.copy()
column_names.append("Class")

df_iris = pd.DataFrame(data=np.c_[iris_data['data'],iris_data['target']], columns=column_names)

#names = ["petal length (cm)", "sepal length (cm)"]
#print(names)

display(df_iris)
#display(column_names)

sns.lmplot(x="sepal length (cm)", y="sepal width (cm)", fit_reg=True, data=df_iris, hue="Class")
sns.lmplot(x="petal length (cm)", y="petal width (cm)", fit_reg=True, data=df_iris, hue="Class")
```

```python
# Class Labelling
# Instead of having classes such as [1,2,3]
# It creates a output of size 3 flags, each flag represents one class
encoder = OneHotEncoder(sparse=False)
y_ = encoder.fit_transform(y)

display(x)
display(y_)

# Data Split
X_train, X_test, Y_train, Y_test = train_test_split(x, y_, test_size=0.20)
```

```python
# Create Model with 3 layers
model = Sequential()
model.add(Input(shape=(4,), name='input'))
model.add(Dense(10, activation='relu', name='fc1'))
model.add(Dense(10, activation='relu', name='fc2'))
model.add(Dense(3, activation='softmax', name='output'))
```

```python
# Adam Optimizer
optmizer = Adam(learning_rate=0.001)
model.compile(optmizer, metrics=['accuracy'] , loss='categorical_crossentropy')
print('Summary of the model')
print(model.summary())
```

```python
# Fit Model
model.fit(X_train, Y_train, verbose=1, batch_size=5, epochs=200)
```

```python
predictions = model.predict(X_test)
print("Predicted Values")
print(predictions)
```

```python
print(Y_test)
```

```python
res = model.evaluate(X_test, Y_test)
print(res)
print('Resulting test set accuracy: {:4f}'.format(res[1]))
```
